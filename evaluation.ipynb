```python
import json
import matplotlib.pyplot as plt
import glob
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
l_path = "models/model_aug/"
```

```python
accs = glob.glob(l_path + "*.json")
len(accs)
```

```python
data_list = list()
for acc in accs:
    with open(acc, "r") as file:
        data = json.load(file)
    if isinstance(data, list):
        data_list.extend(data)
    # Falls die JSON-Datei ein einzelnes Dictionary enth√§lt
    elif isinstance(data, dict):
        data_list.append(data)
```

```python
df = pd.DataFrame(data_list)
```

```python
df
```

```python
plt.figure(figsize=(20, 13))
plt.scatter(np.arange(1, len(df["test_acc"]) + 1), df["test_acc"])
plt.ylim([0, 1])
plt.hlines(y=0.2, xmin=0, xmax=25, colors="C3", label="PulHypStudie_Check_npz_SNR30")
plt.hlines(y=0.2, xmin=26, xmax=39, colors="C4", label="PulHypStudie_Check_npz_SNR20")
plt.hlines(y=0.2, xmin=40, xmax=42, colors="C5", label="PulHypStudie_Check_npz_SNR10")
plt.grid()
plt.legend()
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.show()
```

```python
sorted_idx = np.argsort(df["test_acc"])
plt.figure(figsize=(20, 13))
plt.scatter(np.arange(len(sorted_idx)), df["test_acc"][sorted_idx])
plt.xticks(ticks=np.arange(len(sorted_idx)), labels=np.array(sorted_idx) + 1)
plt.ylim([0, 1])
plt.grid()
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.show()
```
