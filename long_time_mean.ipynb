```python
import numpy as np
import matplotlib.pyplot as plt
from src.data_util import load_preprocess_examples
import glob
import json
import os

import seaborn as sns
import pandas as pd

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
Pigs_all = np.array(["P_{0:02d}_PulHyp".format(i) for i in range(1, 11)])
Pigs_all
```

```python
DAP_mean = dict()
SAP_mean = dict()
MAP_mean = dict()

DAP_var = dict()
SAP_var = dict()
MAP_var = dict()

DAP_std = dict()
SAP_std = dict()
MAP_std = dict()


for pig in Pigs_all:
    X, y, clrs_pig = load_preprocess_examples(
        "/data/PulHypStudie_Check_npz_v2/",
        [pig],
        sap=True,
        get_pig=True,
        shuffle=False,
        norm_eit="block",
        quality_check=False,
    )
    DAP_mean[pig] = np.mean(y[:, 0])
    SAP_mean[pig] = np.mean(y[:, 1])
    MAP_mean[pig] = np.mean(y[:, 2])

    DAP_var[pig] = np.var(y[:, 0])
    SAP_var[pig] = np.var(y[:, 1])
    MAP_var[pig] = np.var(y[:, 2])

    DAP_std[pig] = np.std(y[:, 0])
    SAP_std[pig] = np.std(y[:, 1])
    MAP_std[pig] = np.std(y[:, 2])

    print("Plotting", pig)
    df = pd.DataFrame(y, columns=["DAP", "SAP", "MAP"])
    plt.figure(figsize=(10, 6))
    df.boxplot(column=["DAP", "SAP", "MAP"])
    plt.ylabel("Value")
    plt.xlabel("AP")
    plt.show()
```
